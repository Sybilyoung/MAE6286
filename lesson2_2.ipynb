```python
import numpy
from matplotlib import pyplot
%matplotlib inline
```

```python
def linear_convection(nx,c=1.0,L=2.0,dt=0.025,nt=20):
    # Discretize spatial grid.
    dx = L / (nx - 1)
    x = numpy.linspace(0.0, L, num=nx)
    sigma=dt/dx
    print(sigma)
    # Set initial conditions.
    u0 = numpy.ones(nx)
    mask = numpy.where((x >= 0.5) & (x <= 1.0))
    u0[mask] = 2.0
    # Integrate the solution in time.
    u = u0.copy()
    for n in range(1, nt):
        un = u.copy()
        u[1:] = un[1:] - c * dt / dx * (un[1:] - un[:-1])
    # Plot the solution along with the initial conditions.
    pyplot.figure()
    pyplot.xlabel('x')
    pyplot.ylabel('u')
    pyplot.grid()
    pyplot.plot(x, u0, label='Initial')
    pyplot.plot(x, u, label='Current')
    pyplot.legend()
    pyplot.xlim(0.0, L)
    pyplot.ylim(0.0, 2.5);
```

```python
linear_convection(41)
```
