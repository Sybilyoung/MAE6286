```python
import numpy 
from matplotlib import pyplot
```

```python
#Time
T=100.0
dt=0.01
N=int(T/dt)+1
t=numpy.linspace(0.0,T,num=N)
```

```python
#Initial grid
z0=100.0
b0=10.0
zt=100.0
g=9.81


u=numpy.array([z0,b0])
u[1]
```

```python
z=numpy.zeros(N)
z[0]=z0
b=numpy.zeros(N)
b[0]=b0
```

```python
# Temporal integration using Euler's method.

for n in range(1, N):
    rhs = numpy.array([u[1], g * (1 - u[0] / zt)])
    u = u + dt * rhs
    z[n] = u[0]
```

```python
#plot the solution
from matplotlib import pyplot
%matplotlib inline
pyplot.figure(figsize=(9.0,4.0) ) 
pyplot.title('Elevation of the phugoid over the time')  
pyplot.xlabel('Time [s]')  
pyplot.ylabel('Elevation [m]')  
pyplot.xlim(0.0, 100.0)  
pyplot.ylim(50.0, 150.0)  
pyplot.grid()  
pyplot.plot(t, z, color='b', linestyle='-', linewidth=2);
```

```python
z_exact = (b0 * (zt / g)**0.5 * numpy.sin((g / zt)**0.5 * t) +
           (z0 - zt) * numpy.cos((g / zt)**0.5 * t) + zt)
```

```python
#plot the solution

%matplotlib inline
pyplot.figure(figsize=(10.0,5.0) ) 
pyplot.title('Elevation of the phugoid over the time')  
pyplot.xlabel('Time [s]')  
pyplot.ylabel('Elevation [m]')  
pyplot.xlim(t[0], t[-1])  
pyplot.ylim(50.0, 150.0)  
pyplot.grid()  
pyplot.plot(t, z,label='Numerical', color='C0', linestyle='-', linewidth=2);
pyplot.plot(t, z_exact, label='analytical', color='C1', linestyle='-', linewidth=2);
pyplot.legend();
```
